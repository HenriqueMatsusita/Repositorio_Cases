```python
# Importando biblioteca para manipulação de datas
import datetime
```

```python
# Criando data referência (dia ques esta sendo feito a análise)
data_analise=datetime.date(2019,10,30)
```

```python
# Criando coluna com diferença entre a data de "hoje" e data da última compra
df_churn['dias_inativos']=pd.to_datetime(data_analise)-df_churn['DataUltimaTransacao']
```

```python
# Verificando criação das colunas
display(df_churn)
df_churn.info()
```

```python
# Criando coluna "Cliente_Ativo", caso a última compra do cliente tenha sido a mais de 30 dias, ele será considerado cliente inativo
df_churn['cliente_ativo']=np.where(df_churn['dias_inativos']>datetime.timedelta(days=30),1,0)
# 0 = Ativo
# 1 = Inativo
```

```python
# Criando coluna "Tem_Limite_Credito", caso o limite de mercado seja maior que "0", ele tem limite de crédito
df_churn['tem_limite_credito']=np.where(df_churn['Limite_Credito_Mercado']>0,1,0)
# 0 = não tem limite de credito
# 1 = tem limite de credito
```

```python
# Criando faxias de Idade
# 1º passo - criar lista com os valores que vão ser nosso divisores de faixas
classe_idade = [0,10,20,30,40,50,60,70,100] 
# 2º passo - criar lista com as faixas
labels_idade = ['0 a 10','11 a 20','21 a 30','31 a 40','41 a 50','51 a 60','61 a 70','71 a 100']
# 3º passo - criar coluna no 'df'
df_churn['faixa_etaria']=pd.cut(x=df_churn['Idade'],bins=classe_idade,labels=labels_idade)

# Criando faxias de Score
classe_score = [0,449,549,649,749,850]
labels_score = ['350 a 449','450 a 549','550 a 649','650 a 749','750 a 850']
df_churn['faixa_score']=pd.cut(x=df_churn['Score_Credito'],bins=classe_score,labels=labels_score)

# Criando faxias de Limite de Crédito
classe_limite = [-1,0,500,1000,1500,2000,20000]
labels_limite = ['0','1 a 500','501 a 1000','1001 a 1500','1501 a 2000','+2000']
df_churn['faixa_limite']=pd.cut(x=df_churn['Limite_Credito_Mercado'],bins=classe_limite,labels=labels_limite)

# Criando faxias de Total de pedidos
classe_total_pedido = [0,150,300,450,600,750,900]
labels_total_pedido = ['0 a 150','151 a 300','301 a 450','451 a 600','601 a 750','751 a 900']
df_churn['total_pedido']=pd.cut(x=df_churn['Sum_Pedidos_Acumulados'],bins=classe_total_pedido,labels=labels_total_pedido)

# VIsualizando criação das colunas "Faixas"
display(df_churn)
```

```python
# Verificando Taxa de Churn através da média da coluna "Cliente Ativo"
Tx_Churn = df_churn['cliente_ativo'].mean()
print(Tx_Churn)
```

```python
# Criando tabelas dinâmicas, onde relacionei todas variáveis qualitativas com o fato do cliente ser ativo ou não
df_gb_cred=pd.DataFrame(df_churn.pivot_table(index='Usa_Cartao_Credito',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_estado=pd.DataFrame(df_churn.pivot_table(index='Estado',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_genero=pd.DataFrame(df_churn.pivot_table(index='Gênero',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_tempo=pd.DataFrame(df_churn.pivot_table(index='Tempo_Cliente',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_categoria=pd.DataFrame(df_churn.pivot_table(index='Qte_Categorias',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_fidelidade=pd.DataFrame(df_churn.pivot_table(index='Programa_Fidelidade',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_credito_mercado=pd.DataFrame(df_churn.pivot_table(index='tem_limite_credito',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_faixa_etaria=pd.DataFrame(df_churn.pivot_table(index='faixa_etaria',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_faixa_score=pd.DataFrame(df_churn.pivot_table(index='faixa_score',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_faixa_limite=pd.DataFrame(df_churn.pivot_table(index='faixa_limite',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
df_gb_faixa_total_pedido=pd.DataFrame(df_churn.pivot_table(index='total_pedido',columns='cliente_ativo',aggfunc={'cliente_ativo':'count'}))
```

```python
# Criei uma função para criar as colunas "Frequência Relativa","Frequência Absoluta" e "Churn" nas tabelas dinâmicas
# Variáveis de entrada da função "Variável qualitativa" e o nome do dataframe
def tab_freq(variavel,nome):

    nome['freq_rel_0']=nome[('cliente_ativo', 0)]/nome[('cliente_ativo', 0)].sum()
    nome['freq_rel_1']=nome[('cliente_ativo', 1)]/nome[('cliente_ativo', 1)].sum()
    nome['freq_total_abs']=nome[('cliente_ativo', 1)]+nome[('cliente_ativo', 0)]
    nome['freq_total_rel']=nome['freq_total_abs']/nome['freq_total_abs'].sum()
    nome['Churn']=nome[('cliente_ativo', 1)]/nome['freq_total_abs']
    display(nome)
```

```python
# Executando a função criada 
tab_freq('Usa_Cartao_Credito',df_gb_cred)
tab_freq('Estado', df_gb_estado)
tab_freq('Gênero', df_gb_genero)
tab_freq('Tempo_Cliente', df_gb_tempo)
tab_freq('Qte_Categorias', df_gb_categoria)
tab_freq('Programa_Fidelidade', df_gb_fidelidade)
tab_freq('tem_limite_credito', df_gb_credito_mercado)
tab_freq('faixa_etaria', df_gb_faixa_etaria)
tab_freq('faixa_score', df_gb_faixa_score)
tab_freq('faixa_limite', df_gb_faixa_limite)
tab_freq('total_pedido',df_gb_faixa_total_pedido)
```

```python
# Criando uma coluna "Variavel" com uma string referente a variável em análise
df_gb_cred['Variavel']='Cred'
df_gb_estado['Variavel']='Estado'
df_gb_genero['Variavel']='Genero'
df_gb_tempo['Variavel']='Tempo'
df_gb_categoria['Variavel']='Categoria'
df_gb_fidelidade['Variavel']='Fidelidade'
df_gb_credito_mercado['Variavel']='Tem_Credito'
df_gb_faixa_etaria['Variavel']='Faixa_Estaria'
df_gb_faixa_score['Variavel']='Faixa_Score'
df_gb_faixa_limite['Variavel']='Faixa_Limite'
df_gb_faixa_total_pedido['Variavel']='Faixa_Total_Pedido'
```

```python
# Criando um DF com todas as tabelas dinâmicas criadas e ordenando (DESC) pelo "Churn"
# com objetivo de verificar quais variáveis tem maior tx Churn
df=pd.DataFrame()
df=df.append([df_gb_cred, df_gb_estado, df_gb_genero,df_gb_tempo,df_gb_categoria,df_gb_fidelidade,df_gb_credito_mercado,df_gb_faixa_etaria,df_gb_faixa_score,df_gb_faixa_limite,df_gb_faixa_total_pedido])
df=df.sort_values(by='Churn', ascending=True)
display(df)
```

```python
# Criando Tabela dinâmica para verificar o IV entre as variáveis categoria x Churn
df_gb_categoria.columns
```

```python
# Criando coluna Taxa de clientes inativos (clientes inativos / total clientes) pra cada categoria
df_gb_categoria['Tx_Cliente_Inativo']=df_gb_categoria[( 'cliente_ativo',  1)]/df_gb_categoria[('freq_total_abs', '')]
```

```python
# Criando coluna "ODDS"
df_gb_categoria['Odds']=df_gb_categoria[(    'freq_rel_0', '')]/df_gb_categoria[(    'freq_rel_1', '')]
```

```python
# Criando coluna "LN ODDS"
df_gb_categoria['LNODDS']=np.log(df_gb_categoria['Odds'])
```

```python
# Criar coluna IV
df_gb_categoria['IV']=(df_gb_categoria[(    'freq_rel_0', '')]-df_gb_categoria[(    'freq_rel_1', '')])*df_gb_categoria['LNODDS']
```
