```python
# Importando Bibliotecas
import pandas as pd
import numpy as np
```

```python
# Importando Banco de Dados de Arquivo CSV
df_wl=pd.read_csv('walmart-sales-dataset.csv')
```

```python
# Visualizando Caracteristicas do DataFrame
display(df_wl)
df_wl.info()
```

```python
# Renomeando Colunas
df_wl.rename(columns={'Store':'Loja','Date':'Data','Weekly_Sales':'Vendas_Semanais','Holiday_Flag':'Feriado','Temperature':'Temperatura','Fuel_Price':'Preço_Combustivel','Unemployment':'Desemprego'}, inplace=True)
```

```python
# Corrigindo Tipo da Vaviável "Data"
df_wl['Data']=pd.to_datetime(df_wl['Data'], format='%d-%m-%Y')
```

```python
# Corrigindo Tipo da Vaviável "Feriado"
df_wl['Feriado']=df_wl['Feriado'].astype(str)
```

```python
# Corrigindo Tipo da Vaviável "Feriado"
df_wl['Loja']=df_wl['Loja'].astype(str)
```

```python
# Criando Colunas de Mês e Ano
df_wl['Mês']=df_wl['Data'].dt.strftime('%m')
df_wl['Ano']=df_wl['Data'].dt.strftime('%Y')
```

```python
# Verificando se há valores Null
print('Valores null')
display(df_wl.isnull().sum())

# Verificando qtd de valores únicos
print('Valores Únicos')
display(df_wl.nunique())

# Verificando se há linhas duplicadas
print('Valores duplicados')
df_wl.duplicated().sum()
```

```python
# Transformando Unidade Da Variável Temperatura (Para Celsius)
df_wl['Temperatura']=round((df_wl['Temperatura']-32)*(5/9),2)
```

```python
# Visualização do DataFrame depois do tratamento
display(df_wl)
```

```python
# Medidas Resumo das Variáveis Quantitativas
df_wl.describe()
```

```python
# Importando biblioteca Seaborn
import seaborn as sn
```

```python
# Gráfico de correlação entre variáveis quantitativas (Correlação Pearson)
correlacao_qt=df_wl.corr()
plot=sn.heatmap(correlacao_qt,annot=True,fmt='.1f',linewidths=.6)
# Insight 
# Não é possível tirar grandes conclusões com os resultados obitios, essa inexistência de correlação entre
# as variáveis quantitativas e as vendas semanais, se dá por conta de estarmos analisando todas as lojas
# simultâneamente.
```

```python
# Criando Função Para Calcular o R² (Coeficiente de Determinação)
def calc_r2(var_qual):
    df_r2=pd.DataFrame(df_wl.pivot_table(index=var_qual,values=['Vendas_Semanais'],aggfunc=[ 'count','mean','std']))
    df_r2['var']=pow(df_r2[(  'std', 'Vendas_Semanais')],2)
    df_r2['N_x_Var']=df_r2[('count', 'Vendas_Semanais')]*df_r2['var']
    var_bar=(df_r2['N_x_Var'].sum())/(df_r2[('count', 'Vendas_Semanais')].sum())
    var_total=pow(df_wl['Vendas_Semanais'].std(),2)
    R_quad=1-(var_bar/var_total)
    display(df_r2)
    print('R² =',R_quad)
```

```python
# Aplicando a Função do R², para entender a associação entre as variáveis qualitativas e as vendas semanais
calc_r2('Ano')
calc_r2('Mês')
calc_r2('Feriado')
calc_r2('Loja')
# Insight 
# A variável qualitativa que tem maior associação com as vendas semanais é a variável "Loja"
```

```python
# Criando um Boxplot para visualizar a distribuição das vendas semanais por mês
sn.set(rc = {'figure.figsize':(15,8)})
sn.boxplot(data=df_wl, x="Mês", y="Vendas_Semanais")
# Insight 
# É perceptível que os meses que possuem os melhores dados de vendas semanais são os meses de Novembro e Dezembro
# Os meses novembro e dezembro possuem uma variação grande nos valores entreo 3 quartil e o limite superior, isso ocorre por existir alguma semana
# dentro desses meses que vendas seja bem mais alta que as demais.
```

```python
df_wl.sort_values(by='Vendas_Semanais', ascending=False).head(5)
# Insight
# Os outliers são os valores das vendas semanais da semana que antecede o natal, das lojas que tem as maiores vendas.
# Existe grandes variações ns valores entre o 3 qurtil e o limite superior dos meses Nov e Dez.
```

```python
# Criando um Boxplot para visualizar a distribuição das vendas semanais por mês separando as semanas com feriados
sn.set(rc = {'figure.figsize':(15,8)})
sn.boxplot(data=df_wl, x="Mês", y="Vendas_Semanais", hue="Feriado")
# Insight 
# No mês de Novembro, tem a semana de ações de graça, que é um período que as vendas são mais elevadas, esse periodo é a causa da
# grande variação dos valores do 3 quartil e o limite superior do mês de novembro (boxplot anterior)
# No mês de Dezembro o feriado não é a causa da grande variação nos valores entre o 3 quartil e o limite superior (boxplot anterior)
```

```python
display(df_wl.groupby('Data').agg({'Vendas_Semanais':'sum'}).sort_values(by='Vendas_Semanais',ascending=False).reset_index().head(10))
# Insight
# Dentro do mês de Dezembro, a semana que tem os melhores resultados quanto a vendas, é a semana que antecede o Natal,
# essa semana é a responsável pela grande variação nos valores das vendas apresentadas nos gráficos do Boxplot antereior
```

```python
# Criando tabela dinâmica, agrupando a média das variáveis qualitativas por loja e mês por ano
df_pvt=df_wl.pivot_table(index=['Loja','Mês'], columns=['Ano'], values=['Vendas_Semanais','Desemprego','CPI'], aggfunc={'Desemprego':'mean','CPI':'mean','Vendas_Semanais':'sum'})
```

```python
display(df_pvt)
```

```python
# Criando colunas para analisar a tx de crescimento do CPI, Desemprego e vendas semanais entre 2010 e 2012
df_pvt['TX_CPI-2010-2012']=(df_pvt[(            'CPI', '2012')]/df_pvt[(            'CPI', '2010')]-1)*100
df_pvt['TX_Desemprego-2010-2012']=(df_pvt[(     'Desemprego', '2012')]/df_pvt[(     'Desemprego', '2010')]-1)*100  
df_pvt['$_Vendas-2010-2012']=df_pvt[('Vendas_Semanais', '2012')]-df_pvt[('Vendas_Semanais', '2010')]
```

```python
# Excluindo valores NaN, como não tenho os dados de todos os meses de todos os anos em analise, quando criei as colunas de
# crescimento, foi gerado alguns valores NaN, que para nossa análise não é interessante.
df_pvt2=df_pvt.dropna(axis=0)
```

```python
display(df_pvt2)
```

```python
# Criando tabela dinâmica para agrupar os valores das taxa de crescimento por loja
df_pvt3=df_pvt2.pivot_table(index=['Loja'], values=[(       'TX_CPI-2010-2012',     ''),('TX_Desemprego-2010-2012',     ''),(        'Vendas_Semanais', '2010'),(     '$_Vendas-2010-2012',     '')], aggfunc={(       'TX_CPI-2010-2012',     ''):'mean',('TX_Desemprego-2010-2012',     ''):'mean',(        'Vendas_Semanais', '2010'):'sum',(     '$_Vendas-2010-2012',     ''):'sum'})
```

```python
df_pvt3['TX_Vendas-2010-2012']=df_pvt3[(     '$_Vendas-2010-2012',     '')]/df_pvt3[(        'Vendas_Semanais', '2010')]*100
```

```python
# Ordenando os valores pela tx de crescimento das vendas, afim de verificar quais lojas tiveram os maiores e menores 
# crescimentos.
df_pvt3.sort_values(by=(    'TX_Vendas-2010-2012',     ''), ascending=False).head(45)
```

```python
# Gráfico de correlação entre variáveis quantitativas (Correlação Pearson)
correlacao_pvt3=df_pvt3.corr()
plot=sn.heatmap(correlacao_pvt3,annot=True,fmt='.2f',linewidths=.6)
# Insight
# Tx de desemprego tem uma correlação (moderada - forte) com o aumento das vendas (Quanto maior a redução do desemprego maior 
# o crescimento das lojas)
```

```python
# Análise - Dados das Lojas
df_gp=pd.DataFrame(df_wl.groupby('Loja')['Vendas_Semanais'].describe()).sort_values(by='mean',ascending=False).reset_index()
df_gp['sum']=df_gp['count']*df_gp['mean']
df_gp['Representatividade']=round(df_gp['sum']/df_gp['sum'].sum()*100,2)
df_gp['Acumulado']=df_gp['Representatividade'].cumsum()
display(df_gp.head(6))
display(df_gp.tail(6))
# Insight
# as lojas com maiores e menores médias de vendas semanais
```

### Principais INSIGHTS
- Períodos que possuem as maiores vendas semanais: Meses Novembro e Dezembro, mais epecificamente no feriado de novembro (ação de graças), e na semana que antecede o Natal.
- Dentro das variáveis disponibilizadas, a que mais tem influencia sobre o crescimento das lojas é o desemprego (quanto maior a redução no desemprego, maior será o cresciento das vendas) - COrrelação Moderada (Pearson)
- Lojas que tem as maiores representatividade dentro da franquia são (20, 4, 14, 13, 2 e 10) que represetam 25% das vendas de totais da franquia
